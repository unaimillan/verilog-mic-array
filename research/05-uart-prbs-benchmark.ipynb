```python
from types import SimpleNamespace


def prbs7(code):
    for i in range(8):
        next_bit = ~((code>>6) ^ (code>>5))&0x01
        code = ((code<<1) | next_bit) & 0xFF
    return code


def prbs31(code):
    for i in range(32):
        next_bit = ~((code>>30) ^ (code>>27))&0x01
        code = ((code<<1) | next_bit) & 0xFFFFFFFF
    return code


def generate_prbs7(seed: int, length: int) -> list:
    prbs = [0] * length
    code = seed
    for i in range(length):
        prbs[i] = code & 1  # текущий бит
        # вычисляем следующий бит: XOR от битов, на которые указывает многочлен
        next_bit = ((code >> 6) ^ (code >> 5)) & 1
        code = (code << 1) | next_bit
    return prbs


def generate_prbs31(seed: int, length: int) -> list:
    prbs = [0] * length
    code = seed
    for i in range(length):
        prbs[i] = code & 1  # текущий бит
        # вычисляем следующий бит: XOR от битов, на которые указывает многочлен
        next_bit = ((code >> 30) ^ (code >> 27)) & 1
        code = (code << 1) | next_bit
    return prbs

acc = SimpleNamespace()
acc.x = 0
[[setattr(acc, 'x', prbs7(acc.x)), acc.x][-1] for _ in range(200)]

# generate_prbs7(0, 100)
```
