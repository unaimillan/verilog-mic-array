```python
from collections import defaultdict
from dataclasses import dataclass
from io import StringIO
from pathlib import Path
import re

import numpy as np
import pandas as pd
import plotly.express as plt


DATA_DIR = Path('./data')
DATA_DIR.mkdir(exist_ok=True, parents=True)

RAW_DIR = DATA_DIR / 'raw'
RAW_DIR.mkdir(exist_ok=True)

PROCESSED_DIR = DATA_DIR / 'processed'
PROCESSED_DIR.mkdir(exist_ok=True)
```

```python
spreadsheet_path_list = list(RAW_DIR.glob('*.csv'))
print(len(spreadsheet_path_list), 'record files found')

sp_path = sorted(spreadsheet_path_list, key=lambda x: str(x))[3]
print(f'Processing {sp_path}')
sp_filename = sp_path.name

rec_text_raw = sp_path.read_text()
rec_text = re.sub(r',? +', ',', '\n'.join([raw_line.strip() for raw_line in rec_text_raw.splitlines()]))

# print(rec_text[:1000])

raw_rec_pd = pd.read_csv(StringIO(rec_text), skiprows=0, header=0)
rec_pd = raw_rec_pd['sample']
rec_pd
```

```python
def twos_complement(hexstr, bits):
    value = int(hexstr, 16)
    if value & (1 << (bits - 1)):
        value -= 1 << bits
    return value

twos_complement('FFFFFF', 24)
```

```python
# record = list(map(lambda x: twos_complement(x[:-1], 16), (rec_pd[ 2**13 : ].to_list())))
record = (rec_pd[ 2**13 : ].to_list())

print(record[:100])
plt.line(record)
```

```python
from scipy.ndimage import uniform_filter1d
from scipy.signal import savgol_filter

plt.line(y=[record, savgol_filter(record, window_length=11, polyorder=3), uniform_filter1d(record, size=11)])
```
